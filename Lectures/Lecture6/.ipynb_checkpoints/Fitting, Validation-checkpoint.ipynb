## Validation

Now what is validation, and why is it useful in determining how fit our prediction (or best fit line) might be in modeling our data?

Validation is a strategy that involves separating your dataset to a training set and a test set, training your model on the training set, and testing on the test set. With a strategy like this, you can "train", or find the best fit line, on the training set and see how accurate that model is with your test set. In practice, the way we split our data into the training set and test set is by making **80% of the data the training set and 20% of the data the test set**. These numbers are typically just a rule of thumb, but varying from them slightly does not necessarily make your model more or less accurate.
